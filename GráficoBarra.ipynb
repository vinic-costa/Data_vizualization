## **Visualização de Dados - Gráfico de Barra**

> **RELEMBRANDO A SITUAÇÃO PROBLEMA**

Vamos supor que você seja dono de um banco em Taiwan que oferta serviço de crédito. Você tem em mãos uma tabela em excel com dados dos clientes que contrataram o serviço de crédito. 

`Objetivo`: apresentar um panorama consolidado sobre a situação dos clientes no banco para o restante da empresa.

`Objetivo específico`: analisar a situação dos clientes (valor de crédito concedido, valor pago e extrato), nos últimos meses para o ano de 2005 de acordo com o nível de escolaridade.

```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
# Lendo dados tabulares

credit = pd.read_excel("/content/clientes_credito.xlsx")
```

```python
credit
```

```python
credit_group = credit.groupby(['nivel_escolaridade']).mean().reset_index()
```

```python
# Agrupando dados

credit_group = credit.groupby(['nivel_escolaridade']).mean().reset_index()
credit_group

# 1=Pós-graduação
# 2=Graduação
# 3=Ensino médio
# 4=Outros
# 5=Desconhecido
# 6=Desconhecido
```

```python
# GRÁFICO COM A MATPLOTLIB

plt.figure(figsize=(8,6))
plt.xlabel("Nível de escolaridade")
plt.ylabel("Média de crédito concedido")

plt.title("Valor médio de crédito concedido segundo o nível de escolaridade")

plt.bar(credit_group['nivel_escolaridade'], credit_group['valor_credito'], 
        color = 'green', edgecolor='black', linewidth = 1, align='center');
```

```python
# GRÁFICO COM A MATPLOTLIB

plt.figure(figsize=(8, 6))

plt.xlabel("Nível de escolaridade")
plt.ylabel("Média do valor de crédito concedido")

plt.title("Valor de crédito concedido segundo o nível de escolaridade")

plt.bar(credit_group['nivel_escolaridade'], credit_group['valor_credito'], align='center',
        color='Red', linewidth=1, edgecolor='black',);
```

```python
# GRÁFICO COM A SEABORN

grafico, axes = plt.subplots(2,2, figsize = (20,10))

sns.barplot(data = credit_group, x = 'nivel_escolaridade', y = 'extrato_jun_2005', ax = axes[0,0])
sns.barplot(data = credit_group, x = 'nivel_escolaridade', y = 'extrato_jul_2005', ax = axes[0,1])
sns.barplot(data = credit_group, x = 'nivel_escolaridade', y = 'extrato_ago_2005', ax = axes[1,0])
sns.barplot(data = credit_group, x = 'nivel_escolaridade', y = 'extrato_set_2005', ax = axes[1,1]);
```

```python
# GRÁFICO COM A PLOTLY

fig = go.Figure()

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_abr_2005'],
    marker_color = 'red',
    name = 'Valor pago em 04/2005'))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_mai_2005'],
    marker_color = 'blue',
    name = 'Valor pago em 05/2005'))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_jun_2005'],
    marker_color = 'green',
    name = 'Valor pago em 06/2005'))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_jul_2005'],
    marker_color = 'orange',
    name = 'Valor pago em 07/2005'))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_ago_2005'],
    marker_color = 'crimson',
    name = 'Valor pago em 08/2005'))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_set_2005'],
    marker_color = 'purple',
    name = 'Valor pago em 09/2005'))

fig.update_layout(
    yaxis=dict(
        title='Valor do pagamento em 2005',
        titlefont_size=14,
        tickfont_size=12,),
    
    xaxis=dict(
        title='Nível de escolaridade',
        titlefont_size=14,
        tickfont_size=12,),
    
    legend=dict(
        x=1.0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ))

fig.show()

# Salvando o gráfico interativo
fig.write_html("valor_pago_escolaridade.html")
```
